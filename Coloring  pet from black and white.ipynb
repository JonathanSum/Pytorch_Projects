```
def resize_one(fn, i, path, size):
    dest = path/fn.relative_to(path_hr)
    dest.parent.mkdir(parents=True, exist_ok=True)
    img = PIL.Image.open(fn).convert('LA')
    targ_sz = resize_to(img, size, use_min=True)
    img = img.resize(targ_sz, resample=PIL.Image.BILINEAR).convert('RGB')
    img.save(dest, quality=60)
```

```
def resize_one_testing(fn, i, path, size):
    dest = path/fn.relative_to(path_hr)
    dest.parent.mkdir(parents=True, exist_ok=True)
    img = PIL.Image.open(fn).convert('1')
    # img.show()
    img.save(dest)
```

## Feature loss

```
t = data.valid_ds[0][1].data
t = torch.stack([t,t])
```

```
def gram_matrix(x):
    n,c,h,w = x.size()
    x = x.view(n, c, -1)
    return (x @ x.transpose(1,2))/(c*h*w)
```

```
gram_matrix(t)
```

```
base_loss = F.l1_loss #l1 loss or choose mse loss for checking how close of two pictures are
```

```
vgg_m = vgg16_bn(True).features.cuda().eval()
requires_grad(vgg_m, False)
```

```
blocks = [i-1 for i,o in enumerate(children(vgg_m)) if isinstance(o,nn.MaxPool2d)]
blocks, [vgg_m[i] for i in blocks]
#block is where the layer or features just before the Maxpoo2d which is where before the grid size changes
#second line is check the layer with index i in block
```

```
class FeatureLoss(nn.Module):
    def __init__(self, m_feat, layer_ids, layer_wgts):
        super().__init__()
        self.m_feat = m_feat      #pre-train mode to generate where the loss
        self.loss_features = [self.m_feat[i] for i in layer_ids]       #get the layer for the loss
        self.hooks = hook_outputs(self.loss_features, detach=False)     #use hook to hook the intermedte output
        self.wgts = layer_wgts
        self.metric_names = ['pixel',] + [f'feat_{i}' for i in range(len(layer_ids))
              ] + [f'gram_{i}' for i in range(len(layer_ids))]

    def make_features(self, x, clone=False):
        self.m_feat(x)    #if x is the target, then go thru all of the stored activion(hook)
        return [(o.clone() if clone else o) for o in self.hooks.stored]        #grab the copy of them with hook 
    
    def forward(self, input, target):
        out_feat = self.make_features(target, clone=True)     #get the target activation with hooks and clone
        in_feat = self.make_features(input)           #get the in activation with hooks, no need to clone because...
        self.feat_losses = [base_loss(input,target)]        #step one, compair the pixel loss
        self.feat_losses += [base_loss(f_in, f_out)*w       #step two, compair the actvions in the exit layers    
                             for f_in, f_out, w in zip(in_feat, out_feat, self.wgts)]
        self.feat_losses += [base_loss(gram_matrix(f_in), gram_matrix(f_out))*w**2 * 5e3
                             for f_in, f_out, w in zip(in_feat, out_feat, self.wgts)]
        self.metrics = dict(zip(self.metric_names, self.feat_losses))
        return sum(self.feat_losses)
    
    def __del__(self): self.hooks.remove()
```

```
feat_loss = FeatureLoss(vgg_m, blocks[2:5], [5,15,2])
```

## Train

```
wd = 1e-3
learn = unet_learner(data, arch, wd=wd, loss_func=feat_loss, callback_fns=LossMetrics,
                     blur=True, norm_type=NormType.Weight)
gc.collect();
```

```
learn.lr_find()
learn.recorder.plot()
```

```
lr = 1e-3
```

```
def do_fit(save_name, lrs=slice(lr), pct_start=0.9):
    learn.fit_one_cycle(10, lrs, pct_start=pct_start)
    learn.save(save_name)
    learn.show_results(rows=1, imgsize=5)
```
